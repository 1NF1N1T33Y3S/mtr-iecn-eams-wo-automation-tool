```python
from constants.file_paths import iecc_centralized_log_file_path
from helper.data_helper import DataHelper

data_helper = DataHelper()

df = (
    data_helper
    .set_file_path(iecc_centralized_log_file_path)
    .set_sheet_name("2025")
    .read_excel()
)
len(df)
```

```python
dropped_df = df.dropna(how="all")
```

```python
wo_df = dropped_df.dropna(subset=['Work Order Number'])
len(wo_df)
```

```python
wo_df["fault-cleared"] = wo_df["Fault Cleared"].str.rstrip().str.lstrip().str.lower()
wo_df.head()
```

```python
import pandas as pd
wo_df['report-date'] = pd.to_datetime(wo_df['Fault Report\nDate'])

wo_df
```

```python
wo_df['overdue'] = (pd.Timestamp.today() - wo_df['report-date']).dt.days
wo_df
```

```python
work_order_number_list = wo_df["Work Order Number"].to_list()
```

```python
import re
eams_wo_id_pattern: str = r"5\d{9}$"
i = 0
for e in work_order_number_list:
     e = str(e).lstrip().rstrip().lower()
     is_found = re.search(eams_wo_id_pattern, e)
     if is_found:
        print(f"{i=} {e=} {is_found=}")
        i +=1
```
